```python
ds['rating'].replace({'74 min': 'UR', '84 min': 'UR', '66 min': 'UR', ' ': 'UR'}, inplace = True)
ds.rating.unique()
```

```python
showlist = []
movielist = []

for i in range(len(ds)):
    if ds['type'].iloc[i] == 'TV Show':
        showlist.append(i)
    else:
        movielist.append(i)
        
shows = ds.loc[showlist, :]

movies = ds.loc[movielist, :]

shows = shows.set_index([pd.Index(range(0, len(shows)))])
movies = movies.set_index([pd.Index(range(0, len(movies)))])
```

```python
def theduration(data):
    #count = 0
    duration = []
    
    for i in data:
        if isinstance(i, float):
            duration.append(0)
        else:
            i = i.split(' ')
            duration.append(i[0])
            
    return duration
movies.insert(9, 'length', theduration(movies['duration']))
```

```python
def getdates(dats):
    month_add = []
    
    for i in dats:
        if isinstance (i, float):
            month_add.append(0)
        else:
            i = i.split(' ')
            month_add.append(i[0])
    
    return month_add

movies.insert(5, 'month_add', getdates(movies['date_added']))
```

```python
def theseason(dat):
    dura = []
    
    for i in dat:
        if isinstance (i, float):
            dura.append(0)
        else:
            i = i.split(' ')
            dura.append(i[0])
    
    return dura

shows.insert(9, 'seasons', theseason(shows['duration']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
cat_cols=shows.select_dtypes(include=['object']).columns
num_cols = shows.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(shows[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    shows[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=shows[col])
    plt.show()
```
